# Paso a paso para crear el entorno de trabajo en Jupyter notebook:

1. Necesitamos crear una carpeta de Proyecto en la ruta que deseemos. Por ejemplo: una que se llame 'sample_project_1' en el escritorio.

2. Abrir Anaconda Prompt y situarnos en la carpeta del proyecto:

```bash
    (base) C:\Users\Martin>cd Desktop\sample_project_1
```
```bash
    (base) C:\Users\Martin\Desktop\sample_project_1>
```

3. Luego, crear un entorno personalizado dentro de la carpeta de proyecto:

```bash
    (base) C:\Users\Martin\Desktop\sample_project_1>conda create --prefix ./env pandas numpy matplotlib scikit-learn jupyter
```
   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;darle a yes

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Conda va a descargar y extraer (instalar) todos los paquetes necesarios para la configuración.

4. Después, para activar este entorno, usamos:

```bash
    conda activate C:\Users\Martin\Desktop\sample_project_1\env 
```

5. Finalmente, ejecutamos el comando: jupyter notebook

```bash
    (C:\Users\Martin\Desktop\sample_project_1\env) C:\Users\Martin\Desktop\sample_project_1>jupyter notebook
```

# Paso a paso para subir mis notebooks desde mi PC a GitHub

1. Armar la estructura del directorio en mi PC y ordenarla en conjunto con las notebooks. Por ejemplo, supongamos que\
mi carpeta local es:

```bash
    C:\Users\Martin\Desktop\data-science-practice-2025
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Y está estructurada de esta forma

```text
      data-science-practice-2025/
      ├── 01. Python Essentials/
      └── 02. Intro-to-Pandas/
```

2. Crear el repositorio en GitHub (en el navegador). Por ejemplo:

```bash
    https://github.com/mfuentes3352/data-science-practice-2025
```

3. En la carpeta creada en mi PC, crear un archivo de texto que se llame gitignore y pegar esto:

```bash
    # Entornos virtuales
    env/
    .venv/
        
    # Archivos temporales de Jupyter
    .ipynb_checkpoints/
        
    # Caché de Python
    __pycache__/
    *.pyc
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;guardarlo como '.gitignore' (este paso es importante para evitar subir archivos innecesarios o muy pesados).

4. Abri la terminal cmd. Situarme en el carpeta del proyecto:

```bash
    C:\Users\Martin\Desktop>cd data-science-practice-2025
```
   
5. Inicializar el repositorio local de Git:

```bash
    git init
```

6. Agregar el repositorio remoto (solo la primera vez):

```bash
    git remote add origin https://github.com/mfuentes3352/data-science-practice-2025.git
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;verificarlo con:

```bash
    git remote -v
```
 
7. Agregar todos los archivos al área de staging:

```bash
    git add .
```
  
8. Hacer el primer commit. Por ejemplo:

```bash
    git commit -m "Primer commit: notebooks de Python y Pandas"
```

9. Subir (push) al repositorio remoto en GitHub:

```bash
    git branch -M main
    git push -u origin main
```

10. Si cree una nueva notebook en mi carpeta local (data-science-practice-2025 en este caso) y quiero subir solo ese\
archivo al repositorio en GitHub, tengo que seguir estos pasos en mi terminal:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Supongamos que la nueva notebook se llama (por ejemplo):

```bash
    03. Numpy/numpy_basics.ipynb
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Abrir la terminal y situarme en la carpeta del repositorio:

```bash
    cd "C:\Users\Martin\Desktop\data-science-practice-2025"
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;verificar qué archivos nuevos hay:

```bash
    git status
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Debería ver algo como:

```bash
    Untracked files:
        (use "git add <file>..." to include in what will be committed)
            03. Numpy/numpy_basics.ipynb
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Agregar solo esa notebook nueva:

```bash
    git add "03. Numpy/numpy_basics.ipynb"
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Hacer el commit

```bash
    git commit -m "Agrego notebook de introducción a Numpy"
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Hacer el push al repositorio remoto

```bash
    git push origin main
```
