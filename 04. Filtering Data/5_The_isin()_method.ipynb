```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

## 1. isin(): Single Filtering

```python
# Selecting Apple or HP Laptops
df_laptops['Company'].isin(['Apple','HP'])

# df[<name of the column we want to apply this filter>].isin(<data in the form of a list>)
# "Select the 'Apple' and the 'HP' element from the 'Company' series" 
```

```python
# As we can see here, i got a series with True and False values. if the rows have "Apple"
# or "HP" in the 'Company' feature, it returns True, otherwise it returns False.
```

```python
# Filtering dataframes
df_laptops[df_laptops['Company'].isin(['Apple','HP'])]
```

```python
# And as I can see here, the Company column only has Apple or HP laptops, so the filter works properly.

# let's see how many Apple and how many HP laptops are in this dataframe
df_laptops[df_laptops['Company'].isin(['Apple','HP'])].value_counts('Company')
```

## 2. isin(): Multiple Filtering

```python
# Finding notebooks or ultrabooks from Apple or HP
filter1 = df_laptops['TypeName'].isin(['Notebook','Ultrabook'])
filter2 = df_laptops['Company'].isin(['Apple','HP'])                                     
```

```python
# Filtering dataframe
df_laptops[filter1 & filter2]
```

```python
# And we get here a dataframe that is supposed to have the "Apple" and the "HP"
# in the Company column, and also the values of "Notebook" and "Ultrabook" in the
# TypeName column. And we can see that apparently this conditions are satisfied.
```

```python
df_laptops[filter1 & filter2].value_counts('Company')
```

```python
df_laptops[filter1 & filter2].value_counts('TypeName')
```

```python
# And that's how the isin() method works in Pandas.
```
