## 1. Creating a New Column Based on a Condition using np.where()

```python
import numpy as np
```

Based on the price_euros column, we're going to compare this prices with another price to determine if\
this is either cheap or expensive. So let's use the number 2000. And Let's say that if a laptop costs > 2000,\
this is considered expensive. And if it costs < 2000 is considered cheap.

```python
# creating an array based on price tiers
np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')

# np.where(<condition>,<the value we want to set when this condition is true>,
#          <the value we want to set when this condition is false>)
```

```python
# Now we don't have the True or False values anymore. But now we have the values of cheap
# or expensive
```

```python
# setting it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# showing the first 5 rows
df_laptops.head()
```

```python
# And now if we go to the last column. Here we have a column named 'Price_tier' and it has
# the values of cheap or expensive.

# So for example, in the first row we can see the price we can see the Price in euros is < 2000
# so it is considered Cheap, the tier assigned to this laptop is correct. And then, for example,
# in the 4th row, we can see the price is > 2000, so it's considered expensive, and again, the tier
# assigned to this laptop is correct.
```

```python
# counting values in price_tier column
df_laptops.value_counts('Price_tier')
```

```python
# And we can see that 1166 laptops are considered Cheap and only 137 are considered Expensive.
# And that's how a column can be created based on one condition with the np.where() method.
```

Based on the inches column, we're going to compare the screen size with another screen size to determine if\
the screen is either Big or Small. So let's use the number 15. And Let's say that if a laptop has a screen > 15",\
the screen is considered Big. And if it's < 15 is considered small.

```python
# Creating an array based on the screen size
np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
# setting it to a new column
df_laptops['Screen_size'] = np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
# showing the first 5 columns
df_laptops.head()
```

```python
# And here we have a new column based on the condition (screen_size > 15). So for example, the 1st laptop has 
# a screen of 13.3", it is < 15", so it is considered Small, and in the screen size column we see that it was classified correctly
# whereas, for example, the 3rd laptop has a screen of 15.6", >15", and it is considered Big, and in the screen size column
# we see that it was classified correctly.
```

```python
# counting values in the screen_size column
df_laptops.value_counts('Screen_size')
```

```python
# And as we can see here, there are 835 laptops whose screen size are considered Big
# while 468 are considered Small, because their screens are < 15"
```
