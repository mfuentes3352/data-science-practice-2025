```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

## 1. Rename Column

```python
# renaming column and overwrite values of the dataframe
df_exams = df_exams.rename(columns={'gender':'Gender'})
df_exams
#{'<column_name>':<new_column_name>}
```

```python
# As we can see, the name of the column is Gender now with g in capital letter
```

```python
# renaming column and updating the dataframe with inplace argument
df_exams.rename(columns={'math score': 'MS',	
                         'reading score':'RS',	
                         'writing score':'WS'},inplace=True)
```

```python
# showing dataframe
df_exams
```

```python
# So here I have the df, and as we can see, the three last columns have the values MS, RS, WS. 
# It means they were updated correctly.
```

## 2. Rename Index

```python
# renaming index 0, 1 and 2 and update the dataframe.
# i want the index to be a, b, c, ..., and so on, instead of 0, 1, 2, ...
df_exams.rename(index={0:'A', 1:'B', 2:'C'}, inplace=True)
```

```python
# showing the first 3 rows
df_exams.head(3)
```
