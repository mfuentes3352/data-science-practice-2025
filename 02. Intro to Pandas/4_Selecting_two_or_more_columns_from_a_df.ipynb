```python
# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

### Selecting two or more columns

```python
# for example, i'm going to select the gender and math score columns
df_exams[['gender','math score']]

# we have to use [[ ]] two square brackets to select 2 or + columns
```

```python
# I can see i obtained a df that has 1000 rows and 2 columns
```

```python
# we can verify that this is actually a df by using the type() function
type(df_exams[['gender','math score']])
```

```python
# selecting more that 2 columns using the [[ ]] notation
df_exams[['gender','math score','reading score','writing score']]
```
