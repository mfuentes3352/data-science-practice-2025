```python
import pandas as pd
```

```python
# reading csv file
df = pd.read_csv('players_20.csv')
# setting index
df.set_index('short_name', inplace=True)
# selecting columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# showing dataframe
df.head()
```

# 1. sample()

The sample method helps us extract some random elements from a dataframe and also series.\
And this can help us when we want to analyze a small portion of our dataframe.

```python
# Extracting 10 random elements from the "nationality" column
df['nationality'].sample(10, random_state=42)

# the 'nationality' column contains the countries where the players were born
```

```python
# and here I can see that I got 10 random elements from this df
# 10 soccer players with their respective nationalities
```

```python
# We can control the random number generation with the random_state parameter
```

```python
df.shape
```

```python
# Extracting a random 20% sample of the dataframe
df.sample(frac=0.2, random_state=42)
```

```python
# So here I obtained a 20% portion of this original dataframe
```

```python
18278/5
```

```python
# the original size of the df divided by 5 is 3655.6, so the 20% is approximately 3656 rows.
```

```python
# Let's see how to increase the sample rate which is known as upsampling

# upsample: increase the sampling rate -> frac > 1 (Note: replace parameter has to be True for frac parameter > 1)
# for example if we set frac=2, this means that we want to get a 200%$ of the original df
# in that case we would like to double the number of rows.

df.sample(frac=2, replace=True, random_state=42)
```

```python
# and here we have this df, whose number of rows is 36556, the double than before..

# And that's how random samples can be created with the sample() method.
```
