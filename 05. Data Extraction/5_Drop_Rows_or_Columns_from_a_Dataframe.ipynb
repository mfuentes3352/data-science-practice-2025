```python
import pandas as pd
```

```python
# reading csv file
df = pd.read_csv('players_20.csv')
# setting index
df.set_index('short_name', inplace=True)
# selecting columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# showing dataframe
df.head()
```

# 1. Drop a Row

```python
# Dropping one row
# using the axis parameter
df.drop('L. Messi', axis=0)

# df.drop(index_we_want_to_get_rid_off, axis= 0 (rows) or 1 (columns))
```

```python
# and we get here this df but the row that had the index 'L. Messi' is not here. So we dropped it.
```

```python
# 2nd option: using the index parameter
df.drop(index=['L. Messi'])

# these ways of dropping rows returns a copy of the df but it does not update it! 
```

```python
# drop two or more rows and update the data (inplace=True)
df.drop(['L. Messi', 'Cristiano Ronaldo'], axis=0, inplace=True)
```

```python
df
```

```python
# and we can see that the rows Messi and Ronaldo are gone. The 1st row here is Neymar Jr.
```

# 2. Drop Columns

```python
# Dropping one column
# using the axis parameter
df.drop('long_name', axis=1)
# for example, i want to delete the 'long_name' column. axis=1 represents column
```

```python
# and as we can see here this df starts with the 'age' column. it dropped the 'long:name' column
```

```python
# using index parameter
df.drop(columns=['long_name'])
```

```python
# and as we can see we don't have the 'long_name' column anymore.
```

```python
# Dropping column by position/index (dropping last column)

df.drop(df.columns[[-1]], axis=1)
# the last column 'club' is in the position 6, or -1 is equivalent
```

```python
# so now, as we can see, we don't have the 'club' column anymore. We dropped it.
```

```python
# Dropping two or more columns and update data (inplace=True)
df.drop(['long_name','dob'], axis=1, inplace=True)
```

```python
df
```

```python
# if I print the df I can see that the 'long_name' and 'dob' columns are not here.
```

```python
# And that's how rows and columns can be dropped using the drop() method in Pandas.
```
