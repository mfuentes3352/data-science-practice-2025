```python
import pandas as pd
```

```python
# reading csv file
df = pd.read_csv('players_20.csv')
# setting index
df.set_index('short_name', inplace=True)
# selecting columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# showing dataframe
df.head()
```

# 1. Setting a value to one cell

```python
# Updating L. Messi height
df.loc['L. Messi', 'height_cm'] = 175
```

```python
# Showing dataframe
df.head(3)
```

```python
# we can see that the dataframe has a new value inside the index
# L. Messi and 'height_cm' column. So here it says 175 now
```

# 2. Setting a value to an entire column

```python
# setting height of all players to 190
df.loc[:, 'height_cm'] = 190

# : to indicate we want all the rows
```

```python
# showing dataframe
df.head()
```

```python
# and now we can see that all the data inside the 'height_cm' is 190
```

```python
df.shape
```

# 3. Setting a value to an entire row

```python
# Getting all columns that correspond to player ranked last in fifa
df.iloc[-1, :]
```

```python
# and this is all the data about the player 潘喜明
```

```python
# Setting null values to all columns that corresponds to player ranked last
import numpy as np

df.iloc[-1, :] = np.nan
```

```python
# showing dataframe
df
```

```python
# so in the last row, we can see that the player Pan Ximing	only has null values because we updated it with np.nan
```

# 4. Setting a value to Multiple cells

```python
# Setting a value for all items matching the list of labels

df.loc[['L. Messi','Cristiano Ronaldo'], ['height_cm']] = 175
# i'm going to update the heights of all this players
```

```python
# Showing the dataframe
df.head(5)
```

```python
# and we can see that the height of Messi and Ronaldo is now 175, 
# and the rest is 190 because of the changes we made before.
```

# 5. Setting a value for rows matching a condition

```python
# Setting a value for rows matching a condition
columns = ['age','dob','height_cm','weight_kg']


df.loc[df['height_cm'] > 180, columns] = 150

# df.loc[condition, name_of_columns_we_want_to_obtain]
```

```python
# showing the dataframe
df
```

```python
# so now we can see that all the heights of almost all the players is 150 now
# because their height was above 180. But L. Messi and C. Ronaldo are the exception
# of this rule because we changed their height to 175 (less than 180), so this condition
# was false, and wasn't set to these two heights.
```
