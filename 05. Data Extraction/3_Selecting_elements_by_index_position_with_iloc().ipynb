```python
import pandas as pd
```

```python
# reading csv file
df = pd.read_csv('players_20.csv')
# setting index
df.set_index('short_name', inplace=True)
# selecting columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
# showing dataframe
df.head()
```

# 1. Selecting elements with a single value

iloc[row_position, column_position]

```python
# getting the height of L. Messi
df.iloc[0,3]

# inside the square brackets, instead of writing the name of the index or
# the name of the column we have to write the position. The index of Messi is 0
# and the height is located in the column n° 3 (it starts with 0)
```

```python
# For example, Messi is 170 cm tall
```

```python
# getting the weight of Cristiano Ronaldo
df.iloc[1,4]

# the index of Cristiano Ronaldo is 1 and the 'weight_kg' is located in column n° 4
```

```python
# C. Ronaldo weights 83 kg
```

```python
# getting all the rows inside the 'height_cm' column
df.iloc[:,3]
```

```python
# getting all the columns that corresponds to the index of 'L. Messi'
df.iloc[0,:]
```

```python
# And now we get all the elements (or all the columns) that corresponds to the index L. Messi
```

# 2. Selecting with a list of values

```python
# getting all data about L. Messi and Cristiano Ronaldo
df.iloc[[0, 1]]

# df.iloc[0:2, :] 
```

```python
# getting the height of L. Messi and Cristiano Ronaldo
df.iloc[[0, 1], 3]

# df.iloc[0:2, 3] tambien funciona
```

```python
# getting the height and weight of L. Messi
df.iloc[0, [3, 4]]

# [index_of_messi, [positions_of_height_and_weight]]
# otra opcion posible: df.iloc[0,3:5] 
```

```python
# getting the height and weight of l. Messi and Cristiano Ronaldp
df.iloc[[0, 1], [3, 4]]

# Alternativa: df.iloc[0:2,3:5] 
```

# 3. Selecting a range of data with a slice

start:stop (we should note that the start is included but the stop is excluded).

```python
# Slicing column labels age/club
# players = ['L. Messi', 'Cristiano Ronaldo']
players = [0, 1]

df.iloc[players, 1:7]    # 1:7 represent a slice from the 'age' to the 'club' column
```

```python
# and we get the this slice from the 'age' to the 'club' column and we get the index of Messi and Ronaldo.
```

```python
# Slicing index labels (top1 to top10 player)
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1, 2, 3, 4]

# getting top1 and top10 player name
df.iloc[0:10, columns]
```

```python
# and we get this slice from 0 to 9 and all the values that corresponds
# to the columns from 'age' to 'weight_kg'
```

# 4. Selecting elements with conditions

iloc() cannot accept a boolean Series, but only a boolean list. We have to use the list() function\
to conver a Series into a boolean list.

```python
# one condition: selecting players with height above 180 cm
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1, 2, 3, 4]

df.iloc[list(df['height_cm'] > 180), columns]

# df['height'] > 180 is a boolean series, and we have to convert it to a list
```

```python
# and herew we got this df that has only players with height above 180 cm
```

```python
# multiple conditions: selecting players with height above 180 cm from Argentina

df.iloc[list((df['height_cm'] > 180) & (df['nationality'] == "Argentina")), :]
```

```python
# and here we got all the players with height above 180 cm and these players are from Argentina
```
